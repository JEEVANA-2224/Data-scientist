## Descriptive Statistics

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample dataset: Exam scores of students
np.random.seed(42)
data = {
    "Math_Score": np.random.randint(40, 100, 30),
    "Science_Score": np.random.randint(35, 95, 30)
}
df = pd.DataFrame(data)

print("Dataset:\n", df)
```

```python
# --- Descriptive Statistics ---
desc_stats = df.describe()
print("Descriptive Statistics:\n", desc_stats)
```

```python
# --- Central Tendency Measures ---
central_tendency = {
    "Mean": df.mean(),
    "Median": df.median(),
    "Mode": df.mode().iloc[0]
}
print("\nCentral Tendency:\n", pd.DataFrame(central_tendency))
```

```python
# --- Measures of Dispersion ---
dispersion = {
    "Variance": df.var(),
    "Standard Deviation": df.std(),
    "Range": df.max() - df.min(),
    "IQR": df.quantile(0.75) - df.quantile(0.25)
}
print("\nDispersion Measures:\n", pd.DataFrame(dispersion))
```

```python
# --- Visualization Section ---

# Histogram for Math Score
#plt.figure(figsize=(10, 5))
sns.histplot(df["Math_Score"], bins=6, kde=True, color='yellow', edgecolor='black')
plt.title("Histogram of Math Scores")
plt.xlabel("Score Range")
plt.ylabel("Frequency")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Box Plot for Science Score
plt.figure(figsize=(6, 4))
sns.boxplot(df["Science_Score"], color='pink')
plt.title("Box Plot of Science Scores")
plt.xlabel("Score")
plt.grid(True)
plt.tight_layout()
plt.show()
```
